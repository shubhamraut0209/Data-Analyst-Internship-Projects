# **Step 5: Exploratory Data Analysis (EDA)**

**🧍‍♂️ Gender Analysis**                                                             
                                                                                 
**Objective:-**
To identify which gender has a higher purchasing power.

```python
# Count Plot for Gender Distribution
plt.figure(figsize=(6,4))
sns.countplot(x='Gender', data=df, palette='pastel')
plt.title('Gender Distribution')
plt.show()
```

```python
# Bar Plot – Total Purchase Amount by Gender
plt.figure(figsize=(6,4))
gender_sales = df.groupby('Gender')['Amount'].sum().sort_values(ascending=False)
sns.barplot(x=gender_sales.index, y=gender_sales.values, palette='Set2')
plt.title('Total Purchase Amount by Gender')
plt.ylabel('Total Amount Spent')
plt.show()
```

**Interpretation:-**


*   **Males** contributed a higher total purchase amount compared to females, despite nearly equal participation.
*   This suggests that men tend to spend **more per transaction** on average.

**📊 Age Group Analysis**

**Objective:-**

To understand:

1.   Which age group makes the most purchases
2.   Is there a trend in spending based on age and gender?

```python
# Count Plot – Purchases by Age Group (with Gender Hue)
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Age Group', hue='Gender', palette='coolwarm')
plt.title('Purchase Count by Age Group and Gender')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```

```python
# Bar Plot – Total Purchase Amount by Age Group
plt.figure(figsize=(8,5))
age_sales = df.groupby('Age Group')['Amount'].sum().sort_values(ascending=False)
sns.barplot(x=age_sales.index, y=age_sales.values, palette='viridis')
plt.title('Total Purchase Amount by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Amount Spent')
plt.show()
```

**Interpretation:-**

*   The **26–35 age group** recorded the highest number of purchases and the highest overall spending.
*   Younger adults (18–25) and mid-aged adults (36–45) followed closely behind, indicating strong purchasing behavior in the working-age population

**📍 State Analysis.**

**Objective:-**

To find out:
1.   Which states have the highest number of orders
2.   Which states contribute the most revenue

```python
# Top 10 States by Number of Orders
plt.figure(figsize=(10,6))
top_states_orders = df['State'].value_counts().head(10)
sns.barplot(x=top_states_orders.values, y=top_states_orders.index, palette='cubehelix')
plt.title('Top 10 States by Number of Orders')
plt.xlabel('Number of Orders')
plt.ylabel('State')
plt.show()
```

```python
# Top 10 States by Total Purchase Amount
plt.figure(figsize=(10,6))
top_states_revenue = df.groupby('State')['Amount'].sum().sort_values(ascending=False).head(10)
sns.barplot(x=top_states_revenue.values, y=top_states_revenue.index, palette='Spectral')
plt.title('Top 10 States by Total Revenue')
plt.xlabel('Total Amount Spent')
plt.ylabel('State')
plt.show()
```

**Interpretation:-**

*   **Uttar Pradesh, Maharashtra, and Karnataka** were among the top states in terms of both number of orders and total revenue.
*   This points to a higher customer base or stronger market activity in these regions.

**💍 Marital Status Analysis.**

**Objective:-**
Understand how marital status affects purchasing behavior, and whether there's a difference across genders.

```python
# Count of Orders by Marital Status
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Marital_Status', palette='pastel')
plt.title('Order Count by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Number of Orders')
plt.show()
```

```python
# Total Purchase Amount by Marital Status (Split by Gender)
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='Marital_Status', y='Amount', hue='Gender', estimator=sum, ci=None, palette='Set2')
plt.title('Total Revenue by Marital Status and Gender')
plt.xlabel('Marital Status')
plt.ylabel('Total Amount Spent')
plt.show()
```

**Interpretation:-**

*   **Married individuals** spent more than unmarried ones, especially married men.
*   This could suggest family-related shopping during the New Year season.

**💼 Occupation Analysis**

**Objective:-**
Determine which occupations contribute the most to the number of orders and overall sales.

```python
# Count of Orders by Occupation
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='Occupation', palette='muted', order=df['Occupation'].value_counts().index)
plt.title('Order Count by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.show()
```

```python
# Total Purchase Amount by Occupation
plt.figure(figsize=(10,5))
occupation_sales = df.groupby('Occupation')['Amount'].sum().sort_values(ascending=False).reset_index()

sns.barplot(data=occupation_sales, x='Occupation', y='Amount', palette='Blues_r')
plt.title('Total Sales by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Total Amount Spent')
plt.xticks(rotation=45)
plt.show()
```

**Interpretation:-**

*   **IT Sector Employees and Healthcare Professionals** were the most active shoppers, contributing significantly to overall sales.
*   These professions might have higher disposable incomes or more predictable shopping patterns.

**📦 Product Category Analysis**

**Objective:-**
Understand which product categories are most popular and which generate the highest revenue.

```python
# Count of Orders by Product Category
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='Product_Category', palette='Set2', order=df['Product_Category'].value_counts().index)
plt.title('Order Count by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.show()
```

```python
# Total Purchase Amount by Product Category
plt.figure(figsize=(10,5))
category_sales = df.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False).reset_index()

sns.barplot(data=category_sales, x='Product_Category', y='Amount', palette='Spectral')
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Amount Spent')
plt.xticks(rotation=45)
plt.show()
```

**Interpretation:-**

*   **Clothing & Apparel, Electronics, and Food Products** were the most frequently purchased categories.
*   However, **Electronics and Footwear** brought in the highest total revenue, indicating higher value per item.

**Question 3:-**
Which states generate the highest number of orders and revenue??

```python
# Grouping by State
state_orders = df.groupby('State')['Amount'].agg(['count', 'sum']).reset_index().rename(columns={'count': 'Order_Count', 'sum': 'Total_Revenue'})

# Sorting
top_states = state_orders.sort_values(by='Total_Revenue', ascending=False).head(10)

# Plotting side-by-side bar charts
plt.figure(figsize=(14, 6))

# Total Orders
plt.subplot(1, 2, 1)
sns.barplot(data=top_states, x='State', y='Order_Count', palette='Blues_d')
plt.title('Top 10 States by Number of Orders')
plt.xticks(rotation=45)

# Total Revenue
plt.subplot(1, 2, 2)
sns.barplot(data=top_states, x='State', y='Total_Revenue', palette='Greens_d')
plt.title('Top 10 States by Total Revenue')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

***Answer:-***

*   **Uttar Pradesh** generates the highest number of Orders and revenue.
*   Followed by Maharashtra as the second highest in terms of Orders and revenue.

**Question 5:-**
What is the correlation between age and spending amount, and does this differ by gender?

```python
# Mapping age groups to midpoint values
age_map = {
    '0-17': 8, '18-25': 21, '26-35': 30, '36-45': 40, '46-50': 48,
    '51-55': 53, '55+': 60
}
df['Age_Midpoint'] = df['Age Group'].map(age_map)

# Scatterplot with regression line
sns.lmplot(data=df, x='Age_Midpoint', y='Amount', hue='Gender', aspect=1.5)
plt.title('Correlation between Age and Spending by Gender')
plt.tight_layout()
plt.show()
```

***Answer:-***

### 📊 **Overall Correlation: Age vs Spending**

* The trendlines indicate a **slight positive correlation** between age (Age\_Midpoint) and spending (Amount).
* This means that **as age increases, spending also tends to increase**, though not dramatically.

---

### 🚻 **Gender-Based Differences:**

#### 🔵 Female (F):

* The blue trendline for females shows a **gentle upward slope**.
* Spending slightly increases with age among females.

#### 🟠 Male (M):

* The orange trendline for males is also upward but appears **very similar** in slope to that of females.
* Slight increase in spending with age, just like females.

---

### ✅ **Key Takeaways:**

* **Spending increases modestly with age**, regardless of gender.
* **No significant difference in the correlation pattern** between males and females—both genders follow a nearly identical trend.
* Data points show **wide variability** in spending at each age group, meaning individual spending behavior varies greatly, but the average trend is consistent.
